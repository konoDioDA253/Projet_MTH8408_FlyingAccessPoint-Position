### ComputeFGW function

```julia
using Pkg
Pkg.add("Roots")
using Roots
using LinearAlgebra
using Statistics

function computeFgwPos(x::Vector{Float64}, y::Vector{Float64}, z::Vector{Float64}, SNRi::Vector{Int64}, PT::Int64, fi::Float64, PN::Int64)
    c = 3e8
    N = length(x)
    K = -10 * log10(fi) - 10 * log10(4 * pi / c) - PN - PT

    # Define the function f(pos)
    f(pos) = [norm([x[i], y[i], z[i]] .- pos) - 10^((K - SNRi[i])/20) for i in 1:N]

    # Define the initial guess
    pos0 = [mean(x), mean(y), mean(z)]

    # Use fzero to find the root of f starting from pos0
    pos = fzero(f, pos0)

    return pos
end


Ti = [30, 20, 40]  # packet inter-arrival times (ms)
S = 1  # data rate (Mbps)
Pi = [(20, 30, 40), (50, 60, 70), (80, 90, 100)]  # UAV positions (m)
PMAXT = 1000  # maximum TX power (dBm)
C0i = [10, 20, 30]  # offered load (Mbps)
H = 50  # max. delay (ms)
SNRi = [15, 27, 35]  # signal-to-noise ratio (dB)
fi = 5.8e9  # carrier frequency (Hz)
PN = 0  # noise power (dBm)
x = [Float64(p[1]) for p in Pi]
y = [Float64(p[2]) for p in Pi]
z = [Float64(p[3]) for p in Pi]


fgw_pos = computeFgwPos(x,y,z, SNRi, PMAXT, fi, PN)
println(fgw_pos)
```
